```python
from tkinter import *
from PIL import Image,ImageTk
import time
from tkinter import messagebox

import sqlite3
con=sqlite3.connect(database="banking.sqlite")
table1="""Create table accounts
(AccountNo integer primary key autoincrement,Name text,Password text,
Mobile text,Email text,Balance float,Accountopendate text,Accounttype text)"""

table2="""create table txn(account integer,transactiondate text,transactiontype text,Amount float,
updatedbal float)"""


cur=con.cursor()
try:
    cur.execute(table1)
    cur.execute(table2)
    con.commit()
except Exception as e:
    print(e)



    
    
win=Tk()
win.state("zoomed")
win.configure(bg="Blue")

title=Label(win,text="Banking Automation",bg="Blue",font=('Times new roman',35,'bold','underline'),fg="white")
title.pack()

img=Image.open("logo.jpg")
img=img.resize((100,100))
imgtk=ImageTk.PhotoImage(img,master=win)

logolbl=Label(win,image=imgtk)
logolbl.place(relx=0,rely=0)


def home_screen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="white")
    
    def openaccount():
        frm.destroy()
        openaccount_screen()
        
    def forgetpassword():
        frm.destroy()
        forgetpassword_screen()
        
    def reset():
        acnentry.delete(0,"end")
        pswdentry.delete(0,"end")
        acnentry.focus()
        
    def login():
        acn=acnentry.get()
        pswd=pswdentry.get()
        
        if(len(acn)==0 or len(pswd)==0):
            messagebox.showwarning("validation","feilds can't be empty")
            return
        else:
            con=sqlite3.connect(database="banking.sqlite")
            cur=con.cursor()
            query="select * from accounts where accountno=? and password=?"
            cur.execute(query,(acn,pswd))
            global usertup
            usertup=cur.fetchone()
            print(usertup)
            if(usertup==None):
                messagebox.showerror("Failed","Invalid ACN/Pswd")
                return
            else:
                frm.destroy()
                welcome_frame()
           
           
   
    acnlbl=Label(frm,text="Account No:",bg="white",fg="blue",font=("Times new roman",15,"bold"))
    acnlbl.place(relx=.4,rely=.2)
    
    
    acnentry=Entry(frm,font=("Times new roman",15,"bold"),bd=5)
    acnentry.place(relx=.5,rely=.2)
    acnentry.focus()
    
    pswdlbl=Label(frm,text="Password:",bg="white",fg="blue",font=("Times new roman",15,"bold"))
    pswdlbl.place(relx=.4,rely=.3)
    
    
    pswdentry=Entry(frm,font=("Times new roman",15,"bold"),fg="blue",bd=5,show="*")
    pswdentry.place(relx=.5,rely=.3)
    
    
    loginbtn=Button(frm,command=login,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="login")
    loginbtn.place(relx=.47,rely=.45)
    
    
    resetbtn=Button(frm,command=reset,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Reset")
    resetbtn.place(relx=.54,rely=.45)
    
    
    forgetbtn=Button(frm,command=forgetpassword,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Forget password",width=14)
    forgetbtn.place(relx=.40,rely=.55)
    
    
    openbtn=Button(frm,command=openaccount,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Open New Account",width=14)
    openbtn.place(relx=.55,rely=.55)
    
    
def openaccount_screen(): 
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="white")
    
    
    def back():
        frm.destroy()
        home_screen()
        
    def reset():
        nameentry.delete(0,"end")
        passentry.delete(0,"end")
        mobentry.delete(0,"end")
        emailentry.delete(0,"end")
        nameentry.focus()
          
    def openaccount():
        name=nameentry.get()
        password=passentry.get()
        mob=mobentry.get()
        email=emailentry.get()
        Accounttype="Saving"
        opendate=time.ctime()
        balance=0
        if(len(name)==0 or len(password)==0 or len(mob)==0 or len(email)==0):
            messagebox.showwarning("validation","feilds can't be empty")
            return
        else:
            con=sqlite3.connect(database="banking.sqlite")
            cur=con.cursor()
            query="""insert into accounts(Name,Password,Mobile,Email,Balance,Accountopendate,Accounttype)
            values(?,?,?,?,?,?,?)"""
            cur.execute(query,(name,password,mob,email,balance,opendate,Accounttype))
            con.commit()
            
            cur=con.cursor()
            cur.execute("select max(AccountNo) from accounts")
            tup=cur.fetchone()
            con.close()
            messagebox.showinfo("Success",f"Your Account No. is :{tup[0]}")
            
            
            
            
            
    backbtn=Button(frm,command=back,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Back",width=16)
    backbtn.place(relx=0,rely=0)
    
    frmtitle=Label(frm,text="Open New Account",font=("Times new roman",15,"bold","underline"),bg="white",fg="blue")
    frmtitle.pack()
    
    namelbl=Label(frm,text="Name:",bg="white",fg="blue",font=("Times new roman",15,"bold"))
    namelbl.place(relx=.4,rely=.2)
    
    
    nameentry=Entry(frm,font=("Times new roman",15,"bold"),bd=5)
    nameentry.place(relx=.5,rely=.2)
    
    Passwordlbl=Label(frm,text="Password:",bg="white",fg="blue",font=("Times new roman",15,"bold"))
    Passwordlbl.place(relx=.4,rely=.3)
    
    
    passentry=Entry(frm,font=("Times new roman",15,"bold"),bd=5)
    passentry.place(relx=.5,rely=.3)
    
    
    moblbl=Label(frm,text="Mobile No:",bg="white",fg="blue",font=("Times new roman",15,"bold"))
    moblbl.place(relx=.4,rely=.4)
    
    
    mobentry=Entry(frm,font=("Times new roman",15,"bold"),bd=5)
    mobentry.place(relx=.5,rely=.4)
    
    emaillbl=Label(frm,text="Email:",bg="white",fg="blue",font=("Times new roman",15,"bold"))
    emaillbl.place(relx=.4,rely=.5)
    
    
    emailentry=Entry(frm,font=("Times new roman",15,"bold"),bd=5)
    emailentry.place(relx=.5,rely=.5)
    
    openbtn=Button(frm,command=openaccount,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Open")
    openbtn.place(relx=.5,rely=.6)
    
    resetbtn=Button(frm,command=reset,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Reset")
    resetbtn.place(relx=.58,rely=.6)
    
def forgetpassword_screen(): 
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="white")
    
    
    def back():
        frm.destroy()
        home_screen()
        
        
    def reset():
        acnentry.delete(0,"end")
        mobentry.delete(0,"end")
        emailentry.delete(0,"end")
        acnentry.focus()
        
    def get():
        a=acnentry.get()
        m=mobentry.get()
        e=emailentry.get()
        con=sqlite3.connect(database="banking.sqlite")
        cur=con.cursor()
        query="select password from accounts where accountno=? and mobile=? and email=?"
        cur.execute(query,(a,m,e))
        tup=cur.fetchone()
        if(tup==None):
            messagebox.showerror("Incorrect","Details not found")
        else:
            messagebox.showinfo("Success",f"Your pass is:{tup[0]}")
            frm.destroy()
            home_screen()
    
    backbtn=Button(frm,command=back,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Back",width=16)
    backbtn.place(relx=0,rely=0)
    
    frmtitle=Label(frm,text="Get Your Password",font=("Times new roman",15,"bold","underline"),bg="white",fg="blue")
    frmtitle.pack()
    
    
    acnlbl=Label(frm,text="Account No:",bg="white",fg="blue",font=("Times new roman",15,"bold"))
    acnlbl.place(relx=.4,rely=.2)
    
    
    acnentry=Entry(frm,font=("Times new roman",15,"bold"),bd=5)
    acnentry.place(relx=.5,rely=.2)
    acnentry.focus()
    
    moblbl=Label(frm,text="Mobile No:",bg="white",fg="blue",font=("Times new roman",15,"bold"))
    moblbl.place(relx=.4,rely=.3)
    
    
    mobentry=Entry(frm,font=("Times new roman",15,"bold"),bd=5,fg="blue")
    mobentry.place(relx=.5,rely=.3)
    
    emaillbl=Label(frm,text="Email:",bg="white",fg="blue",font=("Times new roman",15,"bold"))
    emaillbl.place(relx=.4,rely=.4)
    
    
    emailentry=Entry(frm,font=("Times new roman",15,"bold"),bd=5,fg="blue")
    emailentry.place(relx=.5,rely=.4)
    
    getbtn=Button(frm,command=get,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="get")
    getbtn.place(relx=.5,rely=.5)
    
    resetbtn=Button(frm,command=reset,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Reset")
    resetbtn.place(relx=.58,rely=.5)
    
def welcome_frame():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="white")
    
    
    def logout():
        frm.destroy()
        home_screen()
    
    def details():
        nestedfrm=Frame(frm,highlightthickness=1,highlightbackground="black")
        nestedfrm.configure(bg="blue")
        nestedfrm.place(relx=.3,rely=.2,relwidth=.5,relheight=.6)
        
        frmtitle=Label(nestedfrm,text="Your Details",font=("Times new roman",15,"bold","underline"),fg="white",bg="blue")
        frmtitle.pack()

        
        con=sqlite3.connect(database="banking.sqlite")
        cur=con.cursor()
        query="select * from accounts where accountno=?"
        cur.execute(query,(usertup[0],))
        tup=cur.fetchone()
        con.close()
        la=Label(nestedfrm,text=f"Account Number: {tup[0]}",bg='blue',font=("Times new roman",15,"bold"),fg="white")
        la.place(relx=.2,rely=.2)
        
        lb=Label(nestedfrm,text=f"Current Balance: {tup[5]}",bg='blue',font=("Times new roman",15,"bold"),fg="white")
        lb.place(relx=.2,rely=.3)
        
        ldate=Label(nestedfrm,text=f"Account Open Date: {tup[6]}",bg='blue',font=("Times new roman",15,"bold"),fg="white")
        ldate.place(relx=.2,rely=.4)
        
    
    
    def updateprofile():
        nestedfrm=Frame(frm,highlightthickness=1,highlightbackground="black")
        nestedfrm.configure(bg="blue")
        nestedfrm.place(relx=.3,rely=.2,relwidth=.5,relheight=.6)
        
        
        def changepass():
            np=lpe.get()
            con=sqlite3.connect(database="banking.sqlite")
            cur=con.cursor()
            cur.execute("update accounts set password=? where accountno=?",(np,usertup[0]))
            con.commit()
            con.close()
            messagebox.showinfo("Success","Password updated")
            lpe.delete(0,"end")
            return
        
        frmtitle=Label(nestedfrm,text="Update Your Profile",font=("Times new roman",15,"bold","underline"),fg="white",bg="blue")
        frmtitle.pack()
    
        lp=Label(nestedfrm,text="Enter New Password:",bg='blue',font=("Times new roman",15,"bold"),fg="white")
        lp.place(relx=.2,rely=.2)
        
        lpe=Entry(nestedfrm,font=("Times new roman",15,"bold"),bd=5)
        lpe.place(relx=.5,rely=.2)
        
        lpb=Button(nestedfrm,font=("Times new roman",15,"bold"),fg="blue",bg="white",bd=5,text='update',command=changepass)
        lpb.place(relx=.5,rely=.4)
        
    
    def deposit():
        nestedfrm=Frame(frm,highlightthickness=1,highlightbackground="black")
        nestedfrm.configure(bg="blue")
        nestedfrm.place(relx=.3,rely=.2,relwidth=.5,relheight=.6)
        
        
        def updatebal():
            amt=float(dpentry.get())
            con=sqlite3.connect(database="banking.sqlite")
            cur=con.cursor()
            query="update accounts set balance=balance+? where accountno=?"
            cur.execute(query,(amt,usertup[0]))
            con.commit()
            
            cur=con.cursor()
            cur.execute("select balance from accounts where accountno=?",(usertup[0],))
            bal=cur.fetchone()[0]
            
            cur=con.cursor()
            query="insert into txn values(?,?,?,?,?)"
            cur.execute(query,(usertup[0],time.ctime(),"Cr.",amt,bal))
            con.commit()
            con.close()
            
            messagebox.showinfo("Success","Amount deposited")
            return
             
        frmtitle=Label(nestedfrm,text="Deposit Here...",font=("Times new roman",15,"bold","underline"),fg="white",bg="blue")
        frmtitle.pack()
        
        dplbl=Label(nestedfrm,text="Amount:",font=("Times new roman",15,"bold"),fg="white",bg='blue')
        dplbl.place(relx=.2,rely=.2)
        
        dpentry=Entry(nestedfrm,font=('',15,'bold'),bd=5)
        dpentry.place(relx=.35,rely=.2)
        
        dpbtn=Button(nestedfrm,text="deposit",font=('Times new roman',15,'bold'),bd=5,fg="blue",bg="white",command=updatebal)
        dpbtn.place(relx=.35,rely=.4)
        
    
    def withdraw():
        nestedfrm=Frame(frm,highlightthickness=1,highlightbackground="black")
        nestedfrm.configure(bg="blue")
        nestedfrm.place(relx=.3,rely=.2,relwidth=.5,relheight=.6)
        
        frmtitle=Label(nestedfrm,text="Withdraw Here...",font=("Times new roman",15,"bold","underline"),fg="white",bg="blue")
        frmtitle.pack()
        
        def updatebal():
            amt=float(wtentry.get())
            con=sqlite3.connect(database="banking.sqlite")
            cur=con.cursor()
            cur=con.cursor()
            cur.execute("select balance from accounts where accountno=?",(usertup[0],))
            bal=cur.fetchone()[0]
            
            if(bal>=amt):
                query="update accounts set balance=balance-? where accountno=?"
                cur.execute(query,(amt,usertup[0]))
                con.commit()
                cur=con.cursor()
                query="insert into txn values(?,?,?,?,?)"
                cur.execute(query,(usertup[0],time.ctime(),"Cr.",amt,bal))
                con.commit()
                con.close()
            
                messagebox.showinfo("Success","Amount Withdraw")
                return
            else:
                messagebox.showinfo("error","insufficient balance")
        
    
        wtlbl=Label(nestedfrm,text="Amount:",font=("Times new roman",15,"bold"),fg="white",bg='blue')
        wtlbl.place(relx=.2,rely=.2)
        
        wtentry=Entry(nestedfrm,font=('',15,'bold'),bd=5)
        wtentry.place(relx=.35,rely=.2)
        
        wtbtn=Button(nestedfrm,text="Withdraw",font=('Times new roman',15,'bold'),bd=5,fg="blue",bg="white",command=updatebal)
        wtbtn.place(relx=.35,rely=.4)
        
        
        
    
    def txn():
        nestedfrm=Frame(frm,highlightthickness=1,highlightbackground="black")
        nestedfrm.configure(bg="blue")
        nestedfrm.place(relx=.3,rely=.2,relwidth=.5,relheight=.6)
        
        frmtitle=Label(nestedfrm,text="Your Transaction Details",font=("Times new roman",15,"bold","underline"),fg="white",bg="blue")
        frmtitle.pack()
        
        con=sqlite3.connect(database="banking.sqlite")
        cur=con.cursor()
        query="select * from txn where account=?"
        cur.execute(query,(usertup[0],))
        tups=cur.fetchall()
        
        head=Label(nestedfrm,text="\tDate\t\tAmt\tType\tBal",bg='blue',fg="white",font=('Times new roman',15,'bold','underline'))
        head.place(relx=.1,rely=.2)
        
        i=.3
        for tup in tups:
            row=Label(nestedfrm,text=f"{tup[1]}\t\t{tup[3]}\t{tup[2]}\t{tup[4]}",bg='blue',font=('Times new roman',12),fg="white")
            row.place(relx=.1,rely=i)
            i=i+.1
    
    
    
    
    
    logoutbtn=Button(frm,command=logout,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="logout")
    logoutbtn.place(relx=.95,rely=0)
    
    
    welcomelbl=Label(frm,text="Welcome,.......",bg="white",font=("Times new roman",15,"bold"),fg="blue")
    welcomelbl.place(relx=0,rely=0)
    
    timelbl=Label(frm,text="login time:"+time.ctime(),bg="white",font=("Times new roman",15,"bold"),fg="blue")
    timelbl.pack()
    
    detailsbtn=Button(frm,command=details,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="User Details",width=16)
    detailsbtn.place(relx=0,rely=.2)
    
    Updatebtn=Button(frm,command=updateprofile,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Update Account",width=16)
    Updatebtn.place(relx=0,rely=.3)
    
    depositbtn=Button(frm,command=deposit,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Deposit",width=16)
    depositbtn.place(relx=0,rely=.4)
    
    withdrawbtn=Button(frm,command=withdraw,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Withdraw",width=16)
    withdrawbtn.place(relx=0,rely=.5)
    
    txnbtn=Button(frm,command=txn,font=("Times new roman",15,"bold"),fg="white",bg="blue",bd=5,text="Transaction History",width=16)
    txnbtn.place(relx=0,rely=.6)
    
    
    
home_screen()
win.mainloop()
```
